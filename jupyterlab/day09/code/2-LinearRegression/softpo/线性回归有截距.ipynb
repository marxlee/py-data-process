```python
# 如果有截距，求解时，需要梯度下降法求解w和b
```

```python
# 什么是梯度？？？
# 梯度就是导数
# 梯度下降是求解最大值，最小值最快的速度
# 导数是函数变化率最快的方向
```

```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
f = lambda x : (x - 4.3)**2 + 7.5*x + 6
```

```python
# 导数知晓，最小值是0.55
g = lambda x: 2*x - 1.1
```

```python
x = np.linspace(-5,9,100)

plt.plot(x,f(x))
```

```python
# 梯度爆炸
```

```python
result = []
# 设起始值的数值
v_min = np.random.randint(-10,10,size = 1)[0]
result.append(v_min)
# 该值记录梯度下降过程中，上一步的值
v_min_last = v_min + 1
# 什么时候梯度下降结束，精度退出条件
precision = 0.00001
# 在一个退出条件，循环次数
max_time = 3000
# 步幅
step = 10
count = 0
print('-------------------------------随机生成的最小值： ',v_min)
while True:
    if np.abs(v_min - v_min_last) < precision:
        break
    if count > max_time:
        break
    v_min_last = v_min
    v_min = v_min - g(v_min)*step
    result.append(v_min)
    print('-----------------------------梯度下降更新的最小值： ',v_min)
    count += 1
```

```python
plt.figure(figsize=(12,9))
x = np.linspace(-8,5,100)
plt.plot(x,f(x))

# 绘制最小值的点result

plt.scatter(result,f(np.array(result)),marker = '*',color= 'red')
```
