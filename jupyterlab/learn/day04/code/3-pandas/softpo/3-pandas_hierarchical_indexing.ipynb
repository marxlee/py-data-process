### 1) 隐式构造

最常见的方法是给DataFrame构造函数的index参数传递两个或更多的数组

- Series也可以创建多层索引

### 2）DataFrame的操作

(1) 可以直接使用列名称来进行列索引

(2) 使用行索引需要用ix()，loc()等函数

【极其重要】推荐使用loc()函数

## 4. 索引的堆（stack）

- ``stack()``
- ``unstack()``

```python
df2
```

```python
df2.stack(level = 0)
```

```python
# 行变成列
df.unstack(level=[0,1])
```

```python
s = df['Python']
s
```

```python
s.unstack()
```

小技巧】使用stack()的时候，level等于哪一个，哪一个就消失，出现在行里。

【小技巧】使用unstack()的时候，level等于哪一个，哪一个就消失，出现在列里。
