# 处理丢失数据

有两种丢失数据：
- None
- np.nan(NaN)

```python
# 数据清洗
import pandas as pd

from pandas import Series,DataFrame

import numpy as np
```

```python
df = DataFrame(np.random.randint(0,150,size = (150,3)),columns=['Python','Math','En'])
df
```

```python
for i in range(15):
    index = np.random.randint(0,150,size = 1)[0]
    column = np.random.randint(0,3,size = 1)[0]
    df.iloc[index,column] = None
```

```python
df
```

```python
df.info()
```

```python
cond = df.isnull().any(axis = 1)
df[cond]
```

```python
# 非空数据
cond = df.notnull().all(axis = 1)

df[cond]
```

```python
df.dropna()
```

```python
df
```

```python
df.drop(labels=['En'],axis = 1)
```

```python
df
```

```python
cond = df.isnull().any(axis = 1)
index = df[cond].index
index
```

```python
df2 = df.drop(labels=index)
df2
```

```python
cond = (df2 < 60).any(axis = 1)

index = df2[cond].index

df2.drop(labels=index)
```

```python
# 平均分小于60进行过滤
cond = df2.mean(axis = 1) < 90

index = df2[cond].index
df2.drop(index)
```

```python
# 小于60，大于100分
cond1 = df2.mean(axis = 1)< 60 
cond1
```

```python
cond2 = df2.mean(axis = 1) > 100
cond2
```

```python
# 或运算
cond = cond1|cond2

df2[cond]
```

## 1. None

None是Python自带的，其类型为python object。因此，None不能参与到任何计算中。

object类型的运算要比int类型的运算慢得多  
计算不同数据类型求和时间  
%timeit np.arange(1e5,dtype=xxx).sum()

## 2. np.nan（NaN）

np.nan是浮点类型，能参与到计算中。但计算的结果总是NaN。

但可以使用np.nan*()函数来计算nan，此时视nan为0。

```python
df = DataFrame(np.random.randint(0,150,size = (5,3)),columns=['Python','En','Math'])
df
```

```python
df['Python'][0] = None

df['En'][3] = np.NaN
```

```python
df
```

```python
np.NaN + 10
```

```python
None + 10
```

```python
type(np.NaN)
```

## 3. pandas中的None与NaN

### 1) pandas中None与np.nan都视作np.nan

创建DataFrame

使用DataFrame行索引与列索引修改DataFrame数据

### 2) pandas中None与np.nan的操作

- ``isnull()``
- ``notnull()``
- ``dropna()``: 过滤丢失数据
- ``fillna()``: 填充丢失数据

(1)判断函数
- ``isnull()``
- ``notnull()``

(2) 过滤函数
- ``dropna()``

可以选择过滤的是行还是列（默认为行）

也可以选择过滤的方式 how = 'all'

(3) 填充函数 Series/DataFrame
- ``fillna()``

```python
df.fillna(60)
```

```python
df.fillna(value = df.mean())
```

```python
# 中位数，
df.fillna(value=df.median())
```

可以选择前向填充还是后向填充

```python
'''method : {'backfill', 'bfill', 'pad', 'ffill', None}, default None
    Method to use for filling holes in reindexed Series
    pad / ffill: propagate last valid observation forward to next valid
    backfill / bfill: use NEXT valid observation to fill gap'''
df.fillna(method='pad',axis = 1)
```

对于DataFrame来说，还要选择填充的轴axis。记住，对于DataFrame来说：

- axis=0：index/行
- axis=1：columns/列
