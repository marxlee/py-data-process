```python
# PCA降维，通过矩阵运算

import numpy as np
```

```python
import sklearn.datasets as datasets
```

```python
from sklearn.decomposition import PCA
```

```python
iris = datasets.load_iris()
X = iris['data']

y = iris['target']
```

```python
X.shape
```

```python
iris.feature_names
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
s = StandardScaler()
```

```python
pca = PCA(n_components=0.95,whiten=False)

X_pca = pca.fit_transform(X)
s.fit_transform(X_pca)
```

```python
# whiten = True就是对数据进行了归一化standard归一化
pca = PCA(n_components=0.95,whiten=True)

X_pca = pca.fit_transform(X)
X_pca
```

```python
# 1、去中心化
# 2、协方差
# 3、特征值和特征向量计算
# 4、特征值选取百分比大于95%的特征所对应的特征向量
# 5、使用选取出来的特征向量和原数据进行矩阵运算
```

```python
# 1、去中心化
A = X - X.mean(axis = 0)
```

```python
import numpy as np
```

```python
# 协方差
V = np.cov(A,rowvar=False)
V
```

```python
# 3、特征值和特征向量计算
T,TV = np.linalg.eig(V)
display(T,TV)
```

```python
s.fit_transform(X.dot(np.linalg.eigh(V)[-1][:,[2,3]]))
```

```python
# 从数学上看，如果向量v与变换A满足Av=λv，则称向量v是变换A的一个特征向量，λ是相应的特征值。
```

```python
V.dot(TV[:,0])
```

```python
T[0]*TV[:,0]
```

```python
a = np.arange(10)
a.cumsum()
```

```python
T.cumsum()/T.sum()
```

```python
pca = PCA(n_components=0.97,whiten=True)

pca.fit_transform(X)
```

```python
# 前两个特征值，所对应的特征向量，分量97%
P = TV[:,[0,1]]

s.fit_transform(X.dot(P))
```
